```python
%matplotlib inline
import os
from collections import defaultdict
import numpy as np
import matplotlib.pyplot as plt
```

```python
models = ['A', 'B', 'C']
pearsons = [0.1, 0.43, 0.6]
```

```python
plt.figure(figsize=(8, 6))
plt.title("Comparison of best Pearson R of different models", fontsize=15)
plt.plot(models, pearsons, ':o')
plt.xlabel("Models", fontsize=12)
plt.ylabel("Pearson Correlation", fontsize=13.2)
plt.yticks(np.arange(0.0, max(pearsons)+0.3,step=0.2))
plt.tight_layout()
plt.show()
```
